```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eps = 0.01

x = np.arange(0.0, 2 + eps, eps)
y1 = (4 + x) / 3
y2 = 4 * x - 6
```

```python
plt.rc('grid', color='#cccccc')
color = '#4d87ff'
```

```python
fig = plt.figure(figsize = (10, 10))
plt.plot(x, y1, x, y2, color = color) 
plt.fill_between(x, y1, y2, alpha = 0.5, color = color)
plt.axis((0, 2.5, 0, 2.5))
plt.grid(linestyle='dotted')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$', rotation = 0)
plt.savefig('ex1.png')
```

```python
eps = 0.01

x = np.arange(5/11, 4 + eps, eps)
y1 = (x + 1) / 4
y2 = 3 * x - 1
```

```python
fig = plt.figure(figsize = (10, 10))
plt.plot(x, y1, x, y2, color = color) 
plt.fill_between(x, y1, y2, alpha = 0.5, color = color)
plt.axis((0, 2.5, 0, 2.5))
plt.grid(linestyle='dotted')
plt.xlabel(r'$y_1$')
plt.ylabel(r'$y_2$', rotation = 0)
plt.savefig('ex2.png')
```
