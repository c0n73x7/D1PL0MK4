```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eps = 0.01

x = np.arange(0.0, 2 + eps, eps)
y1 = (4 + x) / 3
y2 = 4 * x - 6
```

```python
plt.rc('grid', color='#cccccc')
color = '#4d87ff'
```

```python
fig = plt.figure(figsize = (10, 10))
plt.plot(x, y1, x, y2, color = color) 
plt.fill_between(x, y1, y2, alpha = 0.5, color = color)
plt.axis((0, 2.5, 0, 2.5))
plt.grid(linestyle='dotted')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$', rotation = 0)
plt.savefig('ex1.png')
```

```python
eps = 0.001

x11 = np.arange(0.0, 1 + eps, eps)
x12 = np.sqrt(1/4 - (x11 - 1/2)**2)
```

```python
fig = plt.figure(figsize = (10, 10))
plt.plot(np.hstack([x11[:-1], np.flip(x11)]), np.hstack([x12[:-1], np.flip(-x12)]), color = color)
plt.fill(np.hstack([x11[:-1], np.flip(x11)]), np.hstack([x12[:-1], np.flip(-x12)]), color = color, alpha=0.5)
plt.axis((-0.1, 1.1, -0.6, 0.6))
plt.grid(linestyle='dotted')
plt.xlabel(r'$x_{11}$')
plt.ylabel(r'$x_{12}$', rotation = 0)
plt.savefig('ex3.png')
```
